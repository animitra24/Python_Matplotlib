```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#load dataset using sns
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
x=df["total_bill"]
y=df["tip"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Bill")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.scatter(x,y, c="r")
plt.title("Bill")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c="r")
plt.title("Bill")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c="y", marker="*", s=100, alpha=0.5, linewidths=5, edgecolors="g")#verts="<"
plt.title("Bill")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```
