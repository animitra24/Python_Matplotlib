# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classs = ['STD-1', 'STD-2', 'STD-3', 'STD-4']
Marks_Oral = [7,8,9,10]
Marks_Maths = [65,87,93,100]
Marks_English = [45,65,56,78]
```

```python
plt.bar(classs, Marks_English)
```

```python
#plot horizontal
plt.barh(classs, Marks_English)
```

```python
plt.bar(classs, Marks_English, width = 0.2, align = "edge", color='pink', edgecolor ='g', 
        linewidth = 3, alpha = 0.5, linestyle="--", label='Marks_English')
```

```python
plt.bar(classs, Marks_English, width = 0.2, align = "edge", color='pink', edgecolor ='g', 
        linewidth = 3, alpha = 0.5, linestyle="--", label='Marks_English')
plt.title("Student Mark Sheet", fontsize =20)
plt.xlabel("Class", fontsize=15)
plt.ylabel("Marks_English", fontsize = 18)
plt.show()
```
