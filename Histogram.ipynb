```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ML_Student_Age = np.random.randint(18,45, (50))
PY_Student_Age = np.random.randint(15,40, (50))
```

```python
print(ML_Student_Age)
print(PY_Student_Age)
```

```python
plt.hist(ML_Student_Age)
plt.title("ML_Student_Age")
plt.xlabel("Student Age Category")
plt.ylabel("No. of student")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ML_Student_Age, bins, rwidth=0.7, histtype="bar", orientation='vertical', color='m', label='MLStudent')
plt.hist(PY_Student_Age, bins, rwidth=0.7, histtype="bar", orientation='vertical', color='G', label='PYStudent')
plt.title("ML_Student_Age")
plt.xlabel("Student Age Category")
plt.ylabel("No. of student")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize=(12,7))
plt.hist([ML_Student_Age, PY_Student_Age], bins, rwidth=0.7, histtype="bar", 
         orientation='vertical', color=['R','G'], label=['MLStudent','PYStudent'])
plt.title("ML_Student_Age")
plt.xlabel("Student Age Category")
plt.ylabel("No. of student")
plt.legend()
plt.show()
```

```python
from matplotlib  import style
bins = [15,20,25,30,35,40,45]
style.use("ggplot")
plt.figure(figsize=(12,7))
plt.hist([ML_Student_Age, PY_Student_Age], bins, rwidth=0.7, histtype="bar", 
         orientation='vertical', color=['R','G'], label=['MLStudent','PYStudent'])
plt.title("ML_Student_Age")
plt.xlabel("Student Age Category")
plt.ylabel("No. of student")
plt.legend()
plt.show()
```
